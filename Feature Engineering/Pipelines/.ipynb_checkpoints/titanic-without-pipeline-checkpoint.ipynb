```python
import pandas as pd
import numpy as np
```

```python
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder, OrdinalEncoder
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
import pickle
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('tested.csv', )
df.drop(['PassengerId','Name','Ticket','Cabin'],axis=1,inplace=True)
```

```python
df.info()
```

```python
df.sample(random_state=111)
```

```python
df.isnull().sum()
```

## Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop( 'Survived',axis=1),df['Survived'], random_state=11,test_size=0.3)
X_train.shape
```

## Imupting

```python
age_imputer = SimpleImputer()
X_train_age = age_imputer.fit_transform(X_train[['Age']])
X_test_age = age_imputer.transform(X_test[['Age']])
```

```python
np.isnan(X_test_age).sum()
```

```python
df.sample()
```

# Encoding

```python
OHE = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
```

```python
ohe_train = OHE.fit_transform(X_train[['Sex','Embarked']])
ohe_test = OHE.transform(X_test[['Sex','Embarked']])
```

# Merging

```python
X_train_rem = X_train.drop(columns=['Sex','Age','Embarked'])
X_test_rem = X_test.drop(columns=['Sex','Age','Embarked'])
```

```python
X_train_transformed = np.concatenate((X_train_rem,ohe_train, X_train_age),axis=1)
X_test_transformed = np.concatenate((X_test_rem,ohe_test, X_test_age),axis=1)
```

## Model Training

```python
clf = DecisionTreeClassifier()
clf.fit(X_train_transformed, y_train)
```

```python
y_pred = clf.predict(X_test_transformed)
```

```python
y_pred
```

## Evaluation 

```python
accuracy_score(y_test,y_pred)
```
