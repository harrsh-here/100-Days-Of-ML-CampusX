## Thresholding and filtering brands

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
```

```python
threshold = 100
counts = df['brand'].value_counts()
```

```python
mapped_counts = df['brand'].map(counts)
df['brand'] = np.where(mapped_counts>=threshold,df['brand'],'Others')
```

```python
df.sample(10) # Before count mapping
```

```python
df.sample(10) # AFter count mapping
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,:-1], df['selling_price'])
```
