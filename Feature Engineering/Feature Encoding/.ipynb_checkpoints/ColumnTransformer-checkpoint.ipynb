```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('student_performance_dataset.csv',usecols=['student_id','gender','city_type','mental_state', 'learning_style', 
                                                       'career_goal','final_exam_score','performance_category'])
df.sample(5)
```

```python
df.info()
```

# 1. AAM ZINDAGI

```python
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
```

```python
df['mental_state'].value_counts()
```

```python
cat = [['Burnout','Distracted','Balanced','Focused']]
```

```python
oe = OrdinalEncoder(categories=cat, dtype=np.int32)
oe
```

```python
oe.fit(df[['mental_state']])
```

```python
mental_state_encoded = oe.transform(df[['mental_state']])
mental_state_encoded
```

```python
df.tail()
```

```python
oe = OrdinalEncoder(categories=[['Low','Medium','High']], dtype=np.int32)
```

```python
p_c_encoded  = oe.fit_transform(df[['performance_category']])
p_c_encoded
```

## OHE 

```python
OHE = OneHotEncoder(sparse_output=False, drop= 'first',dtype=np.int32)
OHE
```

```python
ohe_df = df[['gender','city_type','learning_style','career_goal']]
```

```python
OHE.fit(ohe_df)
```

```python
OHE.get_feature_names_out()
```

```python
df['city_type'].unique()
```

```python
pd.DataFrame(OHE.transform(ohe_df),index = ohe_df.index, columns = OHE.get_feature_names_out() )
```

```python
new_df = pd.concat([df['student_id'],
                    pd.DataFrame(OHE.transform(ohe_df),index = ohe_df.index, columns = OHE.get_feature_names_out() ),
                    pd.DataFrame(mental_state_encoded,columns=['mental_state']),
                    df['final_exam_score'],
                    pd.DataFrame(p_c_encoded,columns=['performance_category']),
                    ],axis=1)
new_df
```

```python
df
```

```python
non_cat_c = ['student_id','final_exam_score','mental_state','performance_category']
count = 0
for c in (df.columns):
    if c not in non_cat_c:
        count+= len(df[c].unique())-1
        print(f"{c} : {df[c].unique()}, Count = {count}")
    else :
        count+=1
        print(f"{c} , Count = {count}")
print("\n\n\n")
if new_df.shape[1]==count:
    print(f"Shape Verified!, Count: {count} = {new_df.shape[1] } : Columns")
else: print("Shape Mismatch !!")
```
