## location and type

```python
loc_and_types = soup.find_all('span',class_= 'companyCardWrapper__interLinking')
loc = []
industry = []
loc_count=[]
```

```python
for loc_and_type in loc_and_types:
    parts = loc_and_type.text.strip().split("|")
    if len(parts) == 2:
        industry.append(parts[0])
        locs = parts[1].split("+")
        if len(locs)==2:
            loc.append(locs[0])
            loc_count.append((locs[1].split())[0])
        else:
            loc.append(locs[0])
    else :
        loc.append(None)
        loc_count.append(None)
        industry.append(None)
```

# Data Buidling

```python
pages = 500
df = pd.DataFrame()
```

```python
for i in range(1, pages+1) :  
    print(i)
    url = f'https://www.ambitionbox.com/list-of-companies?campaign=homepage_explore_companies_widget&page={i}'

    #data collection
    headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36'
}
 
    data  = requests.get(url,headers=headers)
    soup = BeautifulSoup(data.text, 'lxml')
    print(data)
    #Company Names
    
    company = soup.find_all('h2')
    len(company)
    #company
    Names = []
    for i in company:
        Names.append(i.text.strip())
    Names = Names[1:-3]
    #len(Names)

    
    #reviews
    all_titles = soup.find_all('span',class_= 'companyCardWrapper__ActionTitle')
    reviews = []
    for title_rating in all_titles:
        if "Reviews" in title_rating:
            temp = (title_rating.parent.find('span',class_='companyCardWrapper__ActionCount').text)
            if "--" not in temp:
                reviews.append(temp)
            else:
                reviews.append("-")

    
    #ratings
    all_ratings = soup.find_all('div',class_= 'rating_text')
    ratings = []
    for rating in all_ratings:
        ratings.append(float(rating.text.strip()))
    #len(ratings)
    #loc and type
    loc_and_types = soup.find_all('span',class_= 'companyCardWrapper__interLinking')
    loc = []
    industry = []
    loc_count=[]
    
    for loc_and_type in loc_and_types:
        parts = loc_and_type.text.strip().split("|")
        if len(parts) == 2:
            industry.append(parts[0])
            locs = parts[1].split("+")
            if len(locs)==2:
                loc.append(locs[0])
                loc_count.append((locs[1].split())[0])
            else:
                loc.append(locs[0])
                loc_count.append(-1)
        else :
            loc.append(-1)
            loc_count.append(-1)
            industry.append("")
        
    
    #Interview Count
    action_titles = soup.find_all('span',class_= 'companyCardWrapper__ActionTitle')
    interview_count = []
    for title in action_titles:
        if "Interviews" in title:
            interview_count.append(title.parent.find('span',class_='companyCardWrapper__ActionCount').text)
    
    data = {
        "CompanyName" : Names,
        "Industry" : industry,
        "Locations" : loc,
        "LocationCount" : loc_count,
        "Rating" : ratings,
        "InterviewCount" : interview_count
    }
    temp_df = pd.DataFrame(data)
    df=pd.concat([df,temp_df],axis=0,ignore_index=True)
    #print(df.head())
    #if i%10:
    clear_output(wait=True)
```
