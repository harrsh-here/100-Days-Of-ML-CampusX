## USING pymysql
### 🔌 SQL Connection Revision
* **The Warning:** Occurs when using a raw DB connection in `pd.read_sql`.
* **The Fix:** Use `sqlalchemy.create_engine`.
* **The URI Format:** `mysql+pymysql://root:password@localhost:3306/db_name`
* **Pandas Command:** `df = pd.read_sql_query("SELECT * FROM table", engine)`

```python
#!pip install pymysql
```

```python
from sqlalchemy import create_engine
import pandas as pd

# 1. Create the engine (replace 'password' and 'dbname' with yours)
# Format: mysql+pymysql://user:password@host:port/dbname
engine = create_engine('mysql+pymysql://root:@localhost:3306/w3schools')

# 2. Use the engine in read_sql_query
df = pd.read_sql_query('SHOW TABLES;', engine)

print(df)
```
