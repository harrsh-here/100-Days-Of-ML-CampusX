From Dict Method
Use from_dict() for more control, like setting orientation ('columns' default or 'index').

text
df = pd.DataFrame.from_dict(data, orient='columns')
For index-oriented dicts (keys as rows), set orient='index'; add columns=['Col1', 'Col2'] for unnamed values.

Simple Key-Value Dict
Convert single-value dicts to list of tuples first.

text
simple_data = {'A': 1, 'B': 2}
df = pd.DataFrame(list(simple_data.items()), columns=['Key', 'Value'])
This creates columns for keys and values.
