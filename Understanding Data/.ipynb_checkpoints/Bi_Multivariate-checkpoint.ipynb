```python
sns.scatterplot(data = tips,x ='total_bill', y= 'tip')  # 
```

### Multivariate

```python
plt.figure(figsize=(8,8))
sns.scatterplot(data = tips,x ='total_bill', y= 'tip', hue = 'sex', palette = 'Set1')  # 
```

```python
plt.figure(figsize=(8,8))
sns.scatterplot(data = tips,x ='total_bill', y= 'tip', hue = 'sex', palette = 'Set1', style='smoker')  # 
```

```python
plt.figure(figsize=(8,8))
sns.scatterplot(data = tips,x ='total_bill', y= 'tip', hue = 'sex', palette = 'Set1', style='smoker', size='size')  # 
```
