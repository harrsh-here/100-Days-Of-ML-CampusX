```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load a built-in dataset (Tips at a restaurant)
df = sns.load_dataset("tips")

plt.figure(figsize=(8, 6))

# 'notch=True' creates the Confidence Interval around the median
sns.boxplot(x='day', y='total_bill', data=df, 
            notch=True, palette="Set2")

# Labels and Title
plt.title('Notched Box Plot: Bill Distribution by Day', fontsize=14)
plt.xlabel('Day of the Week', fontsize=12)
plt.ylabel('Total Bill ($)', fontsize=12)

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset("dots")

plt.figure(figsize=(10, 6))

# 'errorbar="ci"' adds the shaded confidence interval (default is 95%)
sns.lineplot(x='time', y='firing_rate', data=df, 
             errorbar="ci", color="teal", label="Mean Firing Rate")

# Labels and Title
plt.title('Line Plot with 95% Confidence Interval', fontsize=14)
plt.xlabel('Time (ms)', fontsize=12)
plt.ylabel('Firing Rate', fontsize=12)
plt.legend()

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("iris")

plt.figure(figsize=(8, 6))

# Basic boxplot
sns.boxplot(y=df['sepal_length'], color="lightblue")

# Adding manual text labels for understanding
plt.text(0.1, df['sepal_length'].median(), " <-- Median (Middle Value)", color='red', fontweight='bold')
plt.text(0.1, df['sepal_length'].quantile(0.25), " <-- Q1 (Bottom 25%)", color='blue')
plt.text(0.1, df['sepal_length'].quantile(0.75), " <-- Q3 (Top 25%)", color='blue')
plt.text(0.3, 6, "The Box is the IQR\n(Middle 50%)", bbox=dict(facecolor='yellow', alpha=0.3))

plt.title('Detailed Anatomy of a Box Plot', fontsize=14)
plt.ylabel('Sepal Length', fontsize=12)

plt.show()
```
